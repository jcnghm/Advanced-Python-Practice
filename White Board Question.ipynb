```python
# White Board Question

def findNums(n):
    for num in range(1, n):
        if num % 3 == 0 and num % 5 == 0:
            print('FizzBuzz')
        elif num % 3 == 0:
            print('Fizz')
        elif num % 5 == 0:
            print('Buzz')
        else:
            print(num)
            
findNums(20)
        
    
```
